```python
# Load image in RGB format
img = cv2.imread("dog.jpg")  # or any other image
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display original image
plt.figure(figsize=(6,6))
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
gray_img = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(6,6))
plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
inverted_img = 255 - gray_img

plt.figure(figsize=(6,6))
plt.imshow(inverted_img, cmap='gray')
plt.title("Inverted Image")
plt.axis("off")
plt.show()
```

```python
# Gaussian Blur to soften details
blurred = cv2.GaussianBlur(inverted_img, (21, 21), sigmaX=0, sigmaY=0)

plt.figure(figsize=(6,6))
plt.imshow(blurred, cmap='gray')
plt.title("Blurred Inverted Image")
plt.axis("off")
plt.show()
```

```python
# Dodge blend: divide the grayscale image by the inverse of the blurred image
def dodge(front, back):
    result = cv2.divide(front, 255 - back, scale=256)
    return result

pencil_sketch = dodge(gray_img, blurred)

plt.figure(figsize=(6,6))
plt.imshow(pencil_sketch, cmap='gray')
plt.title("Pencil Sketch")
plt.axis("off")
plt.show()
```

```python
cv2.imwrite("pencil_sketch.jpg", pencil_sketch)
```

```python
# Try histogram equalization
equalized = cv2.equalizeHist(pencil_sketch)

plt.figure(figsize=(6,6))
plt.imshow(equalized, cmap='gray')
plt.title("Enhanced Pencil Sketch")
plt.axis("off")
plt.show()
```
